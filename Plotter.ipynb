```python
map_lim = 50
grid_range = 20
N_grids = 2*map_lim // grid_range
map_grid_list = []
N_iter = 100
for row in range(N_grids):
    for col in range(N_grids):
        map_grid_list.append([-map_lim + row*grid_range, -map_lim + (row+1)*grid_range,
                              -map_lim + col*grid_range, -map_lim + (col+1)*grid_range])
```

```python
rng = default_rng()
index_list = rng.choice(len(map_grid_list)**2, size=(N_iter,), replace=False)
for i in index_list:
    agent_ind = i // 25
    enemy_ind = i % 25
    print ("random no: ", i, " agent_ind: ", agent_ind, "enemy_ind: ", enemy_ind)
```
